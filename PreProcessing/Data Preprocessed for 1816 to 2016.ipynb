```python
#In this dataset there are only present from 1916 to 2016.
#In order get the dataset before 1916 and after 2016,we get some other dataset and merge it.
```

```python
movies_1874to2017=pd.read_csv('CSV/movies_metadata.csv')
```

```python
credits=pd.read_csv('CSV/credits.csv')
```

```python
display(movies_1874to2017)
```

```python
movies_1874to2017['release_date']=pd.to_datetime(movies_1874to2017['release_date'], errors='coerce')
movies_1874to2017['year']=movies_1874to2017['release_date'].dt.year
```

```python
movies_1874to2017['year'].value_counts().sort_index()
```

```python
new_data = movies_1874to2017[['genres', 'id', 'title', 'year']].dropna(subset=['year'])
```

```python
new_data['id']=new_data['id'].astype(int)
display(new_data)
```

```python
merged_data=pd.merge(new_data,credits, on="id")
display(merged_data)
```

```python
print(merged_data['genres'][0])
type(merged_data['genres'][0])
```

```python
merged_data['genres']=merged_data['genres'].map(lambda x:ast.literal_eval(x))
merged_data['cast']=merged_data['cast'].map(lambda x:ast.literal_eval(x))
merged_data['crew']=merged_data['crew'].map(lambda x:ast.literal_eval(x))
```

```python
print(merged_data['genres'][0])
type(merged_data['genres'][0])
```

```python
def convert_st(x):
    gen=[]
    s=" "
    for i in x:
        if i.get('name'=='Science Fiction'):
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.NaN
    else:
        return (s.join(gen))
```

```python
merged_data['genrelist']=merged_data['genres'].map(lambda x:convert_st(x))
print(merged_data['genrelist'][0])
```

```python
merged_data['cast'][0]
```

```python
def convert_actor(x,index):
    cast=[]
    s=" "
    for i in x:
        cast.append(i.get('name'))
    if cast==[] or len(cast)<=index:
        return np.NaN
    else:
        return cast[index]
```

```python
merged_data['actor_1_name'] = merged_data['cast'].map(lambda x: convert_actor(x,0))
merged_data['actor_2_name'] = merged_data['cast'].map(lambda x: convert_actor(x,1))
merged_data['actor_3_name'] = merged_data['cast'].map(lambda x: convert_actor(x,2))
```

```python
display(merged_data)
```

```python
print(merged_data['crew'][0])
```

```python
def get_director(x):
    dir=[]
    s=" "
    for i in x:
        if i.get('job')=='Director':
            dir.append(i.get('name'))
    if dir==[]:
        return np.NaN
    else:
        return(s.join(dir))
```

```python
merged_data['director_name']=merged_data['crew'].map(lambda x: get_director(x))
```

```python
display(merged_data)
```

```python
data=merged_data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genrelist','title']]
```

```python
data.isna().sum()
```

```python
data=data.dropna(how='any')
```

```python
data = data.rename(columns={'genrelist':'genres'})
data = data.rename(columns={'title':'movie_title'})
```

```python
data['movie_title'] = data['movie_title'].str.lower()
```

```python
data['comb'] = data['actor_1_name'] + ' ' + data['actor_2_name'] + ' '+ data['actor_3_name'] + ' '+ data['director_name'] +' ' + data['genres']
display(data)
```
