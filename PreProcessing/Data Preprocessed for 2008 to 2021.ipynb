```python
from tmdbv3api import Movie
import requests
import numpy as np
tmdb = TMDb()
tmdb_movie = Movie()

tmdb.api_key = '' # Replace with your actual API key

# def get_genre(x):
#     genres = []
#     result = tmdb_movie.search(x)
    
#     if result['results']:
#         movie_id = result['results'][0]['id']  # Access the 'id' key from the first movie in the 'results' list
#         response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key))
# #         print(response.content)
#         data_json = response.json()
        
#         if data_json['genres']:
#             genres = [genre['name'] for genre in data_json['genres']]
#             return ", ".join(genres)
#     return np.NaN
def get_genre(x):
    try:
        genres = []
        result = tmdb_movie.search(x)
        
        if result['results']:
            movie_id = result['results'][0]['id']
            response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key))
            
            # Check if the request was successful (status_code 200)
            if response.status_code == 200:
                data_json = response.json()
                
                # Check if 'genres' is present in the JSON response
                if 'genres' in data_json:
                    genres = [genre['name'] for genre in data_json['genres']]
                    return ", ".join(genres)
                
        return np.NaN

    except Exception as e:
        print(f"Error: {str(e)}")
        return np.NaN

# get_genre("One Missed Call")
```

```python
movies['genres'] = movies['Title'].map(lambda x: get_genre(str(x)))
```

```python
#Extract the director from the cast and crew dictionary
def extract_director(s):
    if "(director)" in s:
        return s.split("(director)")[0]
    elif "(director/screenplay)" in s:
        return s.split("(director/screenplay)")[0] 
    elif "(directors)" in s:
        return s.split("(directors)")[0]
```

```python
movies['director_name']=movies['Cast and crew'].map(lambda x:extract_director(x))
```

```python
#Extract the actor 1,2 and 3 from the cast and crew dictionary
def extract_actor1(s):
    
    if "screenplay); " in s:
        return ((s.split("screenplay); ")[-1]).split(", ")[0])
    else:
        return "unknown"
```

```python
movies['actor_1_name']=movies['Cast and crew'].map(lambda x:extract_actor1(x))
```

```python
def extract_actor2and3(x,i):
    if len((x.split("screenplay); ")[-1]).split(", ")) < i:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[i-1])
```

```python
movies['actor_2_name']=movies['Cast and crew'].map(lambda x:extract_actor2and3(x,2))
```

```python
movies['actor_3_name']=movies['Cast and crew'].map(lambda x:extract_actor2and3(x,3))
```

```python
movies = movies.rename(columns={'Title': 'movie_title'})
movies['movie_title']=movies['movie_title'].str.lower()
```

```python
new_data = movies.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```
