```python
import pandas as pd
import numpy as np
from tmdbv3api import Movie
from tmdbv3api import TMDb
import requests
```

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2022"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
display(df)
```

```python
column_list = df.columns.tolist()
print(column_list) 
```

```python
movies=df[['Title','Cast and crew']]
display(movies)
```

```python
movies=movies.dropna(how="any")
display(movies)
```

```python
tmdb = TMDb()
tmdb_movie=Movie()
tmdb.api_key = '415fc8ed6895e8c714ed4862143e8882'


def extract_genre(x):
    
    gen=[]
    result=tmdb_movie.search(x)
    if result['results']:
        movieid=result['results'][0]['id']
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movieid, tmdb.api_key))
        data_json= response.json()
        if data_json['genres']:
            gen = [genre['name'] for genre in data_json['genres']]
            return ", ".join(gen)
    return np.NaN
```

```python
movies['genres']=movies['Title'].map(lambda x: extract_genre(x))
```

```python
display(movies)
```

```python
movies.dropna(how="any")
```

```python
def extract_director(s):
    if "(director/screenplay)" in s:
        return s.split("(director/screenplay)")[0]
    elif "(director)" in s:
        return s.split("(director)")[0]
    elif "(directors)" in s:
        return s.split("(directors)")[0]
```

```python
movies['director_name']=movies['Cast and crew'].map(lambda x:extract_director(x))
```

```python
def extract_actor1(s):
    if "screenplay)"in s:
        return ((s.split("screenplay); ")[-1]).split(", ")[0])
    elif "director);"in s:
        return ((s.split("director); ")[-1]).split(", ")[0])
    elif "directors)"in s:
        return ((s.split("directors); ")[-1]).split(", ")[0])
```

```python
movies['actor_1_name']=movies['Cast and crew'].map(lambda x:extract_actor1(x))
```

```python
movies=movies.dropna(how="any")
```

```python
def extract_actor2and3(s,i):
    if len((s.split("screenplay); ")[-1]).split(", ")) < i:
        return np.NaN
    else:
        return ((s.split("screenplay); ")[-1]).split(", ")[i-1])
```

```python
movies['actor_2_name']=movies['Cast and crew'].map(lambda x:extract_actor2and3(x,2))
```

```python
movies['actor_3_name']=movies['Cast and crew'].map(lambda x:extract_actor2and3(x,3))
```

```python
display(movies)
```

```python
movies = movies.rename(columns={'Title': 'movie_title'})
movies['movie_title']=movies['movie_title'].str.lower()
```

```python
new_data = movies.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
display(new_data)
```

```python
new_data['actor_2_name'] = new_data['actor_2_name'].replace(np.nan, 'unknown')
new_data['actor_3_name'] = new_data['actor_3_name'].replace(np.nan, 'unknown')
```

```python
new_data['comb'] = new_data['actor_1_name'] + ' ' + new_data['actor_2_name'] + ' '+ new_data['actor_3_name'] + ' '+ new_data['director_name'] +' '+new_data['genres']
```
